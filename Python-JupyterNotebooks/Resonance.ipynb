## Resonance

For star $M_\star$ , planet $M_{pl}$ and satellite $M_m$

\begin{equation}
    T_{pl}^2  = \frac{4 \pi^2}{G(M_\star+M_{pl})} a_{pl}^3 \\
    T_{m}^2   = \frac{4 \pi^2}{G(M_{pl}+M_{m})} a_m^3 \\
\end{equation}

Mean motion $n = \frac{2\pi}{T}$

Mean motion resonance $MMR$: 
\begin{equation}
\frac{n_m}{n_{pl}} = \frac{T_{pl}}{T_m}
\end{equation}

\begin{equation}
    \frac{n_m}{n_{pl}} = \frac{T_{pl}}{T_m} = \left( \frac{M_{pl}+M_m}{M_\star+M_{pl}}\right)^{1/2} \left( \frac{a_{pl}}{a_m}\right)^{3/2}
\end{equation}

where, $n_m$ is mean motion of satellite around planet and $n_{pl}$ is mean motion of planet around star

```python
a_pl =  1.03            # AU: planet around star 
a_m  =  0.006           # AU: moon around planet
M_m  = 6e24             # Moon: 1 Earth mass
M_pl = 1.16*1.898e27    # Laligurans: 1.17 Jovian mass
M_star =  1.989e30      # Sagarmatha: 1 Solar mass
```

```python
((M_pl+M_m)/(M_star+M_pl))**0.5 * (a_pl/a_m)**1.5
```

~75:1 Resonance between planet around star and moon around planet at 0.006 AU

## Deriving Hinse et al. (2010) formula

Rearranging for Jovian irregular satellites
\begin{equation}
\left( \frac{n_m}{n_{pl}} \right)^{2/3} = \left( \frac{M_{pl}+M_m}{M_\star+M_{pl}}\right)^{1/3}  \left(\frac{a_{pl}}{a_m} \right) \\
\end{equation}

\begin{equation}
\frac{a_{m}}{a_{pl}}  = \left( \frac{n_m}{n_{pl}} \right)^{-2/3} \left( \frac{M_{pl}+M_m}{M_\star+M_{pl}}\right)^{1/3} \\
\end{equation}

\begin{equation}

## Deriving Satyal et al. (2014) formula

For planet $M_{stype}$ in S-type orbit around primary $M_A$ having secondary star $M_B$ as binary:

binary stars ~ star + planet 


planet around one of the binary stars ~ moon

\begin{equation}
M_\star = M_B \\
M_{pl} = M_A \\
M_{m}  = M_{stype} \\
a_{pl} = a_{binary} \\
a_m = a_{stype} \\
\end{equation}

\begin{equation}
    \frac{n_{stype}}{n_{binary}} = \frac{T_{binary}}{T_{stype}} = \left( \frac{M_{A}+M_{stype}}{M_A+M_B}\right)^{1/2} \left( \frac{a_{binary}}{a_{stype}}\right)^{3/2}
\end{equation}

## For Kirkwood gaps: asteroid around Sun and Jupiter around Sun

\begin{equation}
    T_{ast}^2  = \frac{4 \pi^2}{G(M_\odot+M_{ast})} a_{ast}^3 \\
    T_{J}^2   = \frac{4 \pi^2}{G(M_\odot+M_J)} a_J^3 \\
\end{equation}

\begin{equation}
    \frac{n_{ast}}{n_{J}} = \frac{T_{J}}{T_{ast}} = \left( \frac{M_{\odot}+M_{ast}}{M_\odot+M_{J}}\right)^{1/2} \left( \frac{a_{J}}{a_{ast}}\right)^{3/2}
\end{equation}


where, $n_{ast}$ is mean motion of asteroid around Sun and $n_{J}$ is mean motion of Jupiter around Sun

Applying the Kirkwood formula with the asteroid placed at the position of the satellite,

### Pericentre of satellite

\begin{equation}
    a_{sat,peri.} = a_{sat} * (1-e_{sat})
\end{equation}

When the pericentre of satellite reaches the planet,

\begin{equation}
    a_{sat,peri.} = R_p \\
    a_{sat} * (1-e_{sat}) = R_p \\
    1- e_{sat} = \frac{R_p}{a_{sat}} \\
    e_{sat} = 1 - \frac{R_p}{a_{sat}}
\end{equation}

When the pericentre of satellite reaches the Roche limit,

\begin{equation}
    a_{sat,peri.} = d_{Roche} \\
    a_{sat} * (1-e_{sat})= 2.46 R_{pl} \left( \frac{3 \rho_{pl}}{\rho_{sat}} \right)^{1/3} \\
    1-e_{sat} = \frac{ 2.46 R_{pl} \left( \frac{3 \rho_{pl}}{\rho_{sat}} \right)^{1/3}}{a_{sat}} \\
    e_{sat} = 1 - \frac{ 2.46 R_{pl} \left( \frac{3 \rho_{pl}}{\rho_{sat}} \right)^{1/3}}{a_{sat}}
\end{equation}

```python
esat_peri = 1 - Rp/asat
```

```python
Rp/AU
```

```python
rhop   = 1.326    # Density of planet: Jupiter density: g/cm^3
rhosat = 5.52     # Density of satellite: Earth-like : g/cm^3
#rhosat = 3.34    # Density of satellite: Moon-like: g/cm^3
Roche  = 2.46*Rp*(3*rhop/rhosat)**(1/3)    # Roche limit
Roche
```

```python
esat_roche = 1 - Roche/asat
```

### Apocentre of the satellite

\begin{equation}
    a_{sat,apo.} = a_{sat} * (1+e_{sat})
\end{equation}

When the apocentre of the satellite reaches the Hill sphere of the planet,

\begin{equation}
    a_{sat,apo.} = R_H \\
    a_{sat} * (1+e_{sat}) = a_p * (1-e_p) * \left(\frac{m_p+m_{sat}}{3m_\star}\right)^{1/3} \\
\end{equation}

\begin{equation}
    1+e_{sat} = \frac{a_p * (1-e_p) * \left(\frac{m_p+m_{sat}}{3m_\star}\right)^{1/3} }{a_{sat}}
\end{equation}

\begin{equation}
    e_{sat} = \frac{a_p * (1-e_p) * \left(\frac{m_p+m_{sat}}{3m_\star}\right)^{1/3} }{a_{sat}} - 1
\end{equation}

```python
Msat      = 6e24       # Mass of Earth-sized satellite: kg 
#Msat      = 7.34e22    # Mass of Moon-sized satellite: kg
Msun      = 1.989e30   # Mass of Sun: 1 solar mass: kg
Mstar     = 1*Msun     # Mass of Sagarmatha
RH        = ap*(1-ep) * ((Mp+Msat)/(3*Mstar))**(1/3) 
RH/AU
```

```python
esat_apo = RH/asat - 1
```
